```python
from bs4 import BeautifulSoup
import re

def generate_jasper_textfields_from_html_multi_style(converthtml: str) -> str:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(converthtml, 'html.parser')

    # Extract all styles from all style tags
    styles = {}
    for style_content in soup.select('style'):
        if style_content.string:  # Check if style content is not None
            for style_definition in style_content.string.split('}'):
                if '{' in style_definition:
                    parts = style_definition.split('{')
                    if len(parts) == 2:
                        selector, properties = parts
                        selector = selector.strip()
                        properties = properties.strip()
                        styles[selector] = properties

    # Process each span to extract the text and associated styles
    jasper_textfields = ""
    for span in soup.select('span.t'):
        span_id = span.attrs.get('id', '')
        span_text = span.string
        span_class = " ".join(span.attrs.get('class', []))

        # Extract style associated with this span using its id
        style_definition = styles.get(f"#{span_id}", '')
        x_match = re.search(r'left:(\d+)px', style_definition)
        y_match = re.search(r'top:(\d+)px', style_definition)
        x_pos = x_match.group(1) if x_match else print(x_match.group(1))
        y_pos = y_match.group(1) if y_match else print(y_match.group(1))

        # Generate the JasperReport XML for this text span
        jasper_textfields += f"""
            <textField>
                <reportElement style="{span_class}" x="{x_pos}" y="{y_pos}" width="{10 * len(span_text)}" height="30" />
                <textFieldExpression><![CDATA["{span_text}"]]></textFieldExpression>
            </textField>
        """

    return jasper_textfields

htmlToConvert = """<!DOCTYPE html>
<html>
<head>
    <meta http-equiv="X-UA-Compatible" content="IE=Edge" />
    <meta charset="utf-8" />
</head>

<body style="margin: 0;">

<div id="p1" style="overflow: hidden; position: relative; background-color: white; width: 1210px; height: 825px;">

    <!-- Begin shared CSS values -->
    <style class="shared-css" type="text/css" >
        .t {
            transform-origin: bottom left;
            z-index: 2;
            position: absolute;
            white-space: pre;
            overflow: visible;
            line-height: 1.5;
        }
        .text-container {
            white-space: pre;
        }
        @supports (-webkit-touch-callout: none) {
            .text-container {
                white-space: normal;
            }
        }
    </style>
    <!-- End shared CSS values -->


    <!-- Begin inline CSS -->
    <style type="text/css" >

        #t1_1{left:18px;top:21px;letter-spacing:0.14px;}
        #t2_1{left:128px;top:21px;}
        #t3_1{left:165px;top:21px;letter-spacing:0.15px;}
        #t4_1{left:266px;top:21px;letter-spacing:0.15px;}
        #t5_1{left:394px;top:21px;letter-spacing:0.17px;}
        #t6_1{left:18px;top:21px;letter-spacing:0.14px;}
        #t7_1{left:128px;top:21px;}
        #t8_1{left:165px;top:21px;letter-spacing:0.15px;}
        #t9_1{left:266px;top:21px;letter-spacing:0.15px;}
        #ta_1{left:394px;top:21px;letter-spacing:0.16px;}
        #tb_1{left:110px;top:21px;}
        #tc_1{left:156px;top:21px;}
        #td_1{left:101px;top:48px;}
        #te_1{left:697px;top:48px;letter-spacing:0.17px;}
        #tf_1{left:18px;top:48px;letter-spacing:0.17px;}
        #tg_1{left:119px;top:48px;letter-spacing:0.17px;}
        #th_1{left:770px;top:48px;letter-spacing:0.17px;}
        #ti_1{left:770px;top:48px;letter-spacing:0.17px;}
        #tj_1{left:862px;top:48px;}
        #tk_1{left:862px;top:48px;letter-spacing:9.39px;word-spacing:-9.17px;}
        #tl_1{left:889px;top:48px;}
        #tm_1{left:953px;top:48px;letter-spacing:0.17px;}
        #tn_1{left:953px;top:48px;letter-spacing:0.17px;}
        #to_1{left:990px;top:48px;letter-spacing:0.17px;}
        #tp_1{left:770px;top:62px;letter-spacing:0.16px;}
        #tq_1{left:953px;top:62px;letter-spacing:0.11px;}
        #tr_1{left:990px;top:62px;letter-spacing:0.17px;}
        #ts_1{left:266px;top:76px;letter-spacing:0.17px;}
        #tt_1{left:266px;top:76px;letter-spacing:0.17px;}
        #tu_1{left:266px;top:76px;letter-spacing:0.17px;}
        #tv_1{left:358px;top:76px;letter-spacing:0.17px;}
        #tw_1{left:358px;top:76px;letter-spacing:0.17px;}
        #tx_1{left:431px;top:76px;letter-spacing:0.17px;}
        #ty_1{left:431px;top:76px;letter-spacing:3.26px;word-spacing:-3.06px;}
        #tz_1{left:468px;top:76px;letter-spacing:0.17px;}
        #t10_1{left:477px;top:76px;}
        #t11_1{left:422px;top:76px;}
        #t12_1{left:422px;top:76px;}
        #t13_1{left:422px;top:76px;}
        #t14_1{left:788px;top:117px;letter-spacing:0.17px;}
        #t15_1{left:18px;top:131px;letter-spacing:0.15px;}
        #t16_1{left:413px;top:131px;letter-spacing:0.17px;}
        #t17_1{left:165px;top:131px;letter-spacing:0.17px;}
        #t18_1{left:165px;top:131px;letter-spacing:0.15px;}
        #t19_1{left:413px;top:144px;letter-spacing:0.17px;}
        #t1a_1{left:18px;top:144px;letter-spacing:0.17px;}
        #t1b_1{left:541px;top:144px;letter-spacing:0.17px;}
        #t1c_1{left:413px;top:158px;letter-spacing:0.14px;}
        #t1d_1{left:523px;top:158px;}
        #t1e_1{left:18px;top:158px;}
        #t1f_1{left:541px;top:158px;letter-spacing:0.17px;}
        #t1g_1{left:413px;top:172px;letter-spacing:0.14px;}
        #t1h_1{left:523px;top:172px;}
        #t1i_1{left:18px;top:172px;letter-spacing:0.17px;}
        #t1j_1{left:541px;top:172px;letter-spacing:0.17px;}
        #t1k_1{left:413px;top:186px;letter-spacing:0.15px;}
        #t1l_1{left:523px;top:186px;}
        #t1m_1{left:413px;top:199px;letter-spacing:0.17px;}
        #t1n_1{left:18px;top:199px;letter-spacing:0.17px;}
        #t1o_1{left:18px;top:199px;letter-spacing:0.16px;}
        #t1p_1{left:541px;top:199px;letter-spacing:0.17px;}
        #t1q_1{left:18px;top:213px;letter-spacing:0.17px;}
        #t1r_1{left:165px;top:213px;letter-spacing:0.14px;}
        #t1s_1{left:18px;top:227px;letter-spacing:0.17px;}
        #t1t_1{left:110px;top:227px;letter-spacing:0.17px;}
        #t1u_1{left:147px;top:227px;letter-spacing:0.17px;}
        #t1v_1{left:202px;top:227px;letter-spacing:0.16px;}
        #t1w_1{left:633px;top:241px;letter-spacing:0.17px;}
        #t1x_1{left:18px;top:241px;letter-spacing:0.17px;}
        #t1y_1{left:770px;top:241px;letter-spacing:0.17px;}
        #t1z_1{left:770px;top:241px;letter-spacing:0.17px;}
        #t20_1{left:18px;top:254px;letter-spacing:0.15px;}
        #t21_1{left:633px;top:254px;letter-spacing:0.17px;}
        #t22_1{left:147px;top:254px;letter-spacing:0.17px;}
        #t23_1{left:147px;top:254px;letter-spacing:0.17px;}
        #t24_1{left:770px;top:254px;letter-spacing:0.17px;}
        #t25_1{left:18px;top:268px;letter-spacing:0.15px;}
        #t26_1{left:128px;top:268px;}
        #t27_1{left:633px;top:268px;letter-spacing:0.17px;}
        #t28_1{left:147px;top:268px;letter-spacing:0.17px;}
        #t29_1{left:770px;top:268px;letter-spacing:0.17px;}
        #t2a_1{left:18px;top:282px;letter-spacing:0.15px;}
        #t2b_1{left:633px;top:282px;letter-spacing:0.17px;}
        #t2c_1{left:193px;top:282px;letter-spacing:0.17px;}
        #t2d_1{left:770px;top:282px;letter-spacing:0.17px;}
        #t2e_1{left:9px;top:323px;letter-spacing:0.17px;}
        #t2f_1{left:9px;top:323px;letter-spacing:0.17px;}
        #t2g_1{left:9px;top:337px;}
        #t2h_1{left:183px;top:337px;letter-spacing:1.93px;word-spacing:-1.74px;}
        #t2i_1{left:9px;top:351px;}
        #t2j_1{left:183px;top:351px;letter-spacing:1.72px;word-spacing:-1.53px;}
        #t2k_1{left:9px;top:364px;letter-spacing:0.17px;}
        #t2l_1{left:9px;top:364px;letter-spacing:0.17px;}
        #t2m_1{left:9px;top:406px;letter-spacing:0.17px;}
        #t2n_1{left:275px;top:406px;}
        #t2o_1{left:339px;top:406px;}
        #t2p_1{left:403px;top:406px;}
        #t2q_1{left:468px;top:406px;}
        #t2r_1{left:532px;top:406px;}
        #t2s_1{left:596px;top:406px;}
        #t2t_1{left:660px;top:406px;}
        #t2u_1{left:724px;top:406px;}
        #t2v_1{left:788px;top:406px;}
        #t2w_1{left:853px;top:406px;}
        #t2x_1{left:917px;top:406px;}
        #t2y_1{left:981px;top:406px;}
        #t2z_1{left:1045px;top:406px;}
        #t30_1{left:1109px;top:406px;}
        #t31_1{left:1173px;top:406px;}
        #t32_1{left:28px;top:419px;letter-spacing:0.17px;}
        #t33_1{left:238px;top:419px;letter-spacing:0.17px;}
        #t34_1{left:9px;top:433px;letter-spacing:0.17px;}
        #t35_1{left:211px;top:433px;}
        #t36_1{left:275px;top:433px;}
        #t37_1{left:339px;top:433px;}
        #t38_1{left:403px;top:433px;}
        #t39_1{left:28px;top:447px;letter-spacing:0.17px;}
        #t3a_1{left:174px;top:447px;letter-spacing:0.17px;}
        #t3b_1{left:816px;top:461px;letter-spacing:0.17px;}
        #t3c_1{left:1027px;top:461px;letter-spacing:0.17px;}
        #t3d_1{left:1027px;top:461px;letter-spacing:0.19px;word-spacing:9.17px;}

        .s0{font-size:15px;font-family:Courier;color:#000;}
    </style>
    <!-- End inline CSS -->

    <!-- Begin page background -->
    <div id="pg1Overlay" style="width:100%; height:100%; position:absolute; z-index:1; background-color:rgba(0,0,0,0); -webkit-user-select: none;"></div>
    <div id="pg1" style="-webkit-user-select: none;"><object width="1210" height="825" data="1/1.svg" type="image/svg+xml" id="pdf1" style="width:1210px; height:825px; -moz-transform:scale(1); z-index: 0;"></object></div>
    <!-- End page background -->


    <!-- Begin text definitions (Positioned/styled in CSS) -->
    <div class="text-container"><span id="t1_1" class="t s0">Seite: </span><span id="t2_1" class="t s0">( </span><span id="t3_1" class="t s0">) *Kopie* </span><span id="t4_1" class="t s0">LIEFERSCHEIN </span><span id="t5_1" class="t s0">Katag AG - Stralsunder Str.5 - 33605 Bielefeld </span><span id="t6_1" class="t s0">Seite: </span><span id="t7_1" class="t s0">( </span><span id="t8_1" class="t s0">) *Kopie* </span><span id="t9_1" class="t s0">LIEFERSCHEIN </span><span id="ta_1" class="t s0">Katag AG - Stralsunder Str.5 - 33605 Bielefeld </span><span id="tb_1" class="t s0">1 </span><span id="tc_1" class="t s0">1 </span>
        <span id="td_1" class="t s0">/ </span><span id="te_1" class="t s0">Nummer: </span><span id="tf_1" class="t s0">28.09.23 </span><span id="tg_1" class="t s0">15:22 </span><span id="th_1" class="t s0">3034948 </span><span id="ti_1" class="t s0">3034948 </span><span id="tj_1" class="t s0">1 </span><span id="tk_1" class="t s0">1 1 </span><span id="tl_1" class="t s0">1 </span><span id="tm_1" class="t s0">120 </span><span id="tn_1" class="t s0">120 31 </span><span id="to_1" class="t s0">31 </span>
        <span id="tp_1" class="t s0">Auftrag Ind SF </span><span id="tq_1" class="t s0">LO </span><span id="tr_1" class="t s0">Bt </span>
        <span id="ts_1" class="t s0">Liefernr. </span><span id="tt_1" class="t s0">_________ </span><span id="tu_1" class="t s0">Liefernr. 3034948 </span><span id="tv_1" class="t s0">_______ </span><span id="tw_1" class="t s0">3034948 001 </span><span id="tx_1" class="t s0">___ </span><span id="ty_1" class="t s0">001 1 </span><span id="tz_1" class="t s0">__ </span><span id="t10_1" class="t s0">1 </span><span id="t11_1" class="t s0">/ </span><span id="t12_1" class="t s0">_ </span><span id="t13_1" class="t s0">/ </span>
        <span id="t14_1" class="t s0">30349480010112031 </span>
        <span id="t15_1" class="t s0">Kundennummer: </span><span id="t16_1" class="t s0">Streutermin : </span><span id="t17_1" class="t s0">28260001 </span><span id="t18_1" class="t s0">28260001 </span>
        <span id="t19_1" class="t s0">Auftragsdat.: </span><span id="t1a_1" class="t s0">Young Fashion Behrendt </span><span id="t1b_1" class="t s0">28.09.23 </span>
        <span id="t1c_1" class="t s0">Marke </span><span id="t1d_1" class="t s0">: </span><span id="t1e_1" class="t s0">. </span><span id="t1f_1" class="t s0">50 ***Mc Percy </span>
        <span id="t1g_1" class="t s0">Thema </span><span id="t1h_1" class="t s0">: </span><span id="t1i_1" class="t s0">Holmpassage, Holm 39 </span><span id="t1j_1" class="t s0">KATAGABRUF Katag Abrufe </span>
        <span id="t1k_1" class="t s0">Prospekt </span><span id="t1l_1" class="t s0">: </span>
        <span id="t1m_1" class="t s0">Auftragstyp : </span><span id="t1n_1" class="t s0">D-24937 Flensburg </span><span id="t1o_1" class="t s0">D-24937 Flensburg </span><span id="t1p_1" class="t s0">291 Katag Lagerauftrag Anschlusshaus </span>
        <span id="t1q_1" class="t s0">Verdichtung...: VERSA </span><span id="t1r_1" class="t s0">VERSA </span>
        <span id="t1s_1" class="t s0">Versandart</span><span id="t1t_1" class="t s0">....</span><span id="t1u_1" class="t s0">: 200 Hängend: DKS / Liegend: DPD-unfr. </span><span id="t1v_1" class="t s0">Hängend: DKS / Liegend: DPD-unfr. </span>
        <span id="t1w_1" class="t s0">Kundentermin : </span><span id="t1x_1" class="t s0">Denim Jeans </span><span id="t1y_1" class="t s0">28.09.23 </span><span id="t1z_1" class="t s0">28.09.23 </span>
        <span id="t20_1" class="t s0">Modellnummer: </span><span id="t21_1" class="t s0">Kundenauftrag: </span><span id="t22_1" class="t s0">210000943 </span><span id="t23_1" class="t s0">210000943 </span><span id="t24_1" class="t s0">3034948 </span>
        <span id="t25_1" class="t s0">Lagerplatz </span><span id="t26_1" class="t s0">: </span><span id="t27_1" class="t s0">Art.Nr. Lief.: </span><span id="t28_1" class="t s0">P10 </span><span id="t29_1" class="t s0">700 NOS </span>
        <span id="t2a_1" class="t s0">Bereich: </span><span id="t2b_1" class="t s0">Art.Bez.Lief.: </span><span id="t2c_1" class="t s0">2100 Katag, HAKA I / Hosen </span><span id="t2d_1" class="t s0">605 5 P. Regular </span>
        <span id="t2e_1" class="t s0">================================================================================================================================== </span><span id="t2f_1" class="t s0">================================================================================================================================== </span>
        <span id="t2g_1" class="t s0">A </span><span id="t2h_1" class="t s0">31/30 32/30 33/30 34/30 36/30 38/30 40/30 31/32 32/32 33/32 34/32 36/32 38/32 40/32 31/34 32/34 </span>
        <span id="t2i_1" class="t s0">B </span><span id="t2j_1" class="t s0">33/34 34/34 36/34 38/34 40/34 </span>
        <span id="t2k_1" class="t s0">================================================================================================================================== </span><span id="t2l_1" class="t s0">================================================================================================================================== </span>
        <span id="t2m_1" class="t s0">A BLUE STONE </span><span id="t2n_1" class="t s0">1 </span><span id="t2o_1" class="t s0">1 </span><span id="t2p_1" class="t s0">1 </span><span id="t2q_1" class="t s0">1 </span><span id="t2r_1" class="t s0">1 </span><span id="t2s_1" class="t s0">1 </span><span id="t2t_1" class="t s0">1 </span><span id="t2u_1" class="t s0">1 </span><span id="t2v_1" class="t s0">1 </span><span id="t2w_1" class="t s0">1 </span><span id="t2x_1" class="t s0">1 </span><span id="t2y_1" class="t s0">1 </span><span id="t2z_1" class="t s0">1 </span><span id="t30_1" class="t s0">1 </span><span id="t31_1" class="t s0">1 </span>
        <span id="t32_1" class="t s0">600 </span><span id="t33_1" class="t s0">______ ______ ______ ______ ______ ______ ______ ______ ______ ______ ______ ______ ______ ______ ______ </span>
        <span id="t34_1" class="t s0">B BLUE STONE </span><span id="t35_1" class="t s0">1 </span><span id="t36_1" class="t s0">1 </span><span id="t37_1" class="t s0">1 </span><span id="t38_1" class="t s0">1 </span>
        <span id="t39_1" class="t s0">600 </span><span id="t3a_1" class="t s0">______ ______ ______ ______ </span>
        <span id="t3b_1" class="t s0">Stück gesamt : </span><span id="t3c_1" class="t s0">19 </span><span id="t3d_1" class="t s0">19 STK </span></div>
    <!-- End text definitions -->


</div>
</body>
</html>

"""
print(generate_jasper_textfields_from_html_multi_style(htmlToConvert))
```

```python
import re

def transform_jasper_report(xml_content: str) -> str:
    added_parameters = []
    added_fields = []

    skip_all = False  # global variable to check if the user has chosen the "Skip All" option

    # Diese Funktion ersetzt das textField anhand der Auswahl des Benutzers
    def replace_text_field(match):
        nonlocal skip_all
        # Wenn skip_all True ist, geben Sie das textField so zurück, wie es ist
        if skip_all:
            return match.group(0)

        # textField-Inhalt extrahieren
        text_field_content = match.group(0)

        # Fragen Sie den Benutzer nach einer Wahl
        print("Für das textField:", text_field_content)
        choice = input("Wählen Sie eine Aktion (1: Parameter [default], 2: Feld, 3: Überspringen, 4: Alles überspringen): ")

        # Wenn keine Auswahl getroffen wird, wird "Parameter" als Standardwert festgelegt
        if choice.strip() == '':
            choice = '1'

        # Wenn der Benutzer "Alles überspringen" wählt
        if choice == '4':
            skip_all = True
            return text_field_content

        # Wenn der Benutzer überspringt
        if choice == '3':
            return text_field_content

        # Extrahieren Sie den String aus textFieldExpression
        expression_match = re.search(r'<textFieldExpression><!\[CDATA\[(.*?)\]\]></textFieldExpression>', text_field_content)
        if not expression_match:
            return text_field_content
        extracted_string = expression_match.group(1)

        # Fragen Sie den Benutzer nach dem Namen
        name = input(f"Geben Sie einen Namen für den {'Parameter' if choice == '1' else 'Feld'} ein: ")

        # Fragen Sie nach dem Datentyp
        data_type = input(f"Wählen Sie den Datentyp (1: java.lang.String [default], 2: java.lang.Integer): ")
        if data_type.strip() == '' or data_type == '1':
            data_type = 'java.lang.String'
        else:
            data_type = 'java.lang.Integer'

        # Erstellen Sie Parameter oder Feld basierend auf der Wahl des Benutzers
        if choice == '1':
            param_str = (
                f'<parameter name="{name}" class="{data_type}">\n'
                f'\t<defaultValueExpression><![CDATA[{extracted_string}]]></defaultValueExpression>\n'
                f'</parameter>'
            )
            added_parameters.append(param_str)
        else:
            field_str = f'<field name="{name}" class="{data_type}">\n\t<defaultValueExpression><![CDATA[{extracted_string}]]></defaultValueExpression>\n</field>'
            added_fields.append(field_str)

        # Erstellen Sie den Ersatzstring
        prefix = 'P' if choice == '1' else 'F'
        replacement = f'<textFieldExpression><![CDATA[${prefix}{{{name}}}]]></textFieldExpression>'

        # Ersetzen Sie den Inhalt mit re.sub()
        new_text_field_content = re.sub(r'<textFieldExpression><!\[CDATA\[(.*?)\]\]></textFieldExpression>', replacement, text_field_content)

        return new_text_field_content

    # Ersetzen Sie alle textFields im XML-Inhalt
    modified_xml = re.sub(r'(<textField>.*?</textField>)', replace_text_field, xml_content, flags=re.DOTALL)

    # Finden Sie den Einfügepunkt für die Parameter
    params_insertion_point = modified_xml.rfind('</parameter>')
    if params_insertion_point != -1:
        modified_xml = modified_xml[:params_insertion_point + 12] + "\n\n" + "\n".join(added_parameters) + modified_xml[params_insertion_point + 12:]
    else:
        modified_xml += "\n\n<!-- Added Parameter Definitions -->\n"
        modified_xml += "\n".join(added_parameters)

    # Finden Sie den Einfügepunkt für die Felder
    fields_insertion_point = re.search(r'</queryString>', modified_xml)
    if fields_insertion_point:
        modified_xml = modified_xml[:fields_insertion_point.end()] + "\n\n" + "\n".join(added_fields) + modified_xml[fields_insertion_point.end():]
    else:
        modified_xml += "\n\n<!-- Added Field Definitions -->\n"
        modified_xml += "\n".join(added_fields)

    return modified_xml

# Let's load the provided XML and test the function
xml_content = """
    <?xml version="1.0" encoding="UTF-8"?>
<!-- Created with Jaspersoft Studio version 6.20.6.final using JasperReports Library version 6.20.6-5c96b6aa8a39ac1dc6b6bea4b81168e16dd39231  -->
<jasperReport xmlns="http://jasperreports.sourceforge.net/jasperreports" xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" xsi:schemaLocation="http://jasperreports.sourceforge.net/jasperreports http://jasperreports.sourceforge.net/xsd/jasperreport.xsd" name="B2B_DeliveryNote" pageWidth="1286" pageHeight="1819" columnWidth="555" leftMargin="20" rightMargin="20" topMargin="20" bottomMargin="20" isFloatColumnFooter="true" uuid="8e05fc2d-d534-416e-9828-f0a2760e7a7f">
	<property name="com.jaspersoft.studio.data.sql.tables" value=""/>
	<property name="com.jaspersoft.studio.data.defaultdataadapter" value="jespersoft_report_Db"/>
	<style name="t s0" fontName="Arial Unicode MS" fontSize="13"/>
	<style name="t s1" fontName="Arial Unicode MS" fontSize="12"/>
	<style name="t s2" fontName="Arial" fontSize="17" isBold="true"/>
	<style name="t s3" fontName="Arial Unicode MS" fontSize="11"/>
	<style name="t s4" fontName="Arial" fontSize="12" isBold="true"/>
	<style name="t s5" fontName="Arial Unicode MS" fontSize="8"/>
	<style name="t s6" fontName="Arial" fontSize="24" isBold="true"/>
	<style name="t s7" fontName="Arial Unicode MS" fontSize="15"/>
	<style name="t s8" fontName="Arial" fontSize="55" isBold="true"/>
	<parameter name="Header" class="java.lang.String">
		<defaultValueExpression><![CDATA["WRSTBHVR "]]></defaultValueExpression>
	</parameter>
	<queryString language="SQL">
		<![CDATA[Select * From jespersoft_report_data.ArticleDataRows]]>
	</queryString>
	<field name="id" class="java.lang.Integer">
		<property name="com.jaspersoft.studio.field.name" value="id"/>
		<property name="com.jaspersoft.studio.field.label" value="id"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="article_id" class="java.lang.Integer">
		<property name="com.jaspersoft.studio.field.name" value="article_id"/>
		<property name="com.jaspersoft.studio.field.label" value="article_id"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="row_num" class="java.lang.Integer">
		<property name="com.jaspersoft.studio.field.name" value="row_num"/>
		<property name="com.jaspersoft.studio.field.label" value="row_num"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="row_name" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="row_name"/>
		<property name="com.jaspersoft.studio.field.label" value="row_name"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="row_data_values_sum" class="java.lang.Integer">
		<property name="com.jaspersoft.studio.field.name" value="row_data_values_sum"/>
		<property name="com.jaspersoft.studio.field.label" value="row_data_values_sum"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="data_count" class="java.lang.Integer">
		<property name="com.jaspersoft.studio.field.name" value="data_count"/>
		<property name="com.jaspersoft.studio.field.label" value="data_count"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value0" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value0"/>
		<property name="com.jaspersoft.studio.field.label" value="value0"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value1" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value1"/>
		<property name="com.jaspersoft.studio.field.label" value="value1"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value2" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value2"/>
		<property name="com.jaspersoft.studio.field.label" value="value2"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value3" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value3"/>
		<property name="com.jaspersoft.studio.field.label" value="value3"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value4" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value4"/>
		<property name="com.jaspersoft.studio.field.label" value="value4"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value5" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value5"/>
		<property name="com.jaspersoft.studio.field.label" value="value5"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value6" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value6"/>
		<property name="com.jaspersoft.studio.field.label" value="value6"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value7" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value7"/>
		<property name="com.jaspersoft.studio.field.label" value="value7"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value8" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value8"/>
		<property name="com.jaspersoft.studio.field.label" value="value8"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value9" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value9"/>
		<property name="com.jaspersoft.studio.field.label" value="value9"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value10" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value10"/>
		<property name="com.jaspersoft.studio.field.label" value="value10"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value11" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value11"/>
		<property name="com.jaspersoft.studio.field.label" value="value11"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<field name="value12" class="java.lang.String">
		<property name="com.jaspersoft.studio.field.name" value="value12"/>
		<property name="com.jaspersoft.studio.field.label" value="value12"/>
		<property name="com.jaspersoft.studio.field.tree.path" value="ArticleDataRows"/>
	</field>
	<group name="id">
		<groupExpression><![CDATA[$F{id}]]></groupExpression>
	</group>
	<background>
		<band splitType="Stretch"/>
	</background>
	<title>
		<band splitType="Stretch"/>
	</title>
	<pageHeader>
		<band height="450" splitType="Stretch">
			<textField>
				<reportElement style="t s8" x="290" y="143" width="416" height="77" uuid="a80924d8-085b-4624-a9c1-f65974565537"/>
				<textFieldExpression><![CDATA[$P{Header}]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="24" y="234" width="630" height="30" uuid="324cbcb0-200c-4bae-a152-ba96901db464"/>
				<textFieldExpression><![CDATA["Worst Behavior GmbH - Carl-Benz-Straße 17 - 89597 Munderkingen "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="695" y="267" width="60" height="30" uuid="ba6e7f98-70b2-46ae-af16-dbbd2064368b"/>
				<textFieldExpression><![CDATA["64128 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="24" y="267" width="300" height="30" uuid="62179801-77db-4644-ad39-889c36532c7c"/>
				<textFieldExpression><![CDATA["Zalando OpCo Polska Sp. z o.o "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="554" y="267" width="190" height="30" uuid="3706f7d1-97a6-4b62-bd20-fe4ee25db15d"/>
				<textFieldExpression><![CDATA["Lieferscheinnummer "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="24" y="284" width="300" height="30" uuid="40db5959-d6ae-4fec-8c59-e8bd648f7749"/>
				<textFieldExpression><![CDATA["Zalando OpCo Polska Sp. z o.o "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="24" y="302" width="180" height="30" uuid="6a53cccf-23d4-413a-82c5-6c73555494b0"/>
				<textFieldExpression><![CDATA["ul. Innowacyina 1 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="554" y="302" width="60" height="30" uuid="70573571-151c-4bd8-bf1d-c364a77267fb"/>
				<textFieldExpression><![CDATA["Datum "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="695" y="302" width="110" height="30" uuid="3be76a89-81e1-417d-9ad2-015bf5b73056"/>
				<textFieldExpression><![CDATA["19.10.2023 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="24" y="317" width="150" height="30" uuid="64254185-eea4-4e40-81df-339ff887c153"/>
				<textFieldExpression><![CDATA["95-080 Gluchow "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="695" y="317" width="60" height="30" uuid="68b054be-5c5d-437b-bf07-c28de66c2b0c"/>
				<textFieldExpression><![CDATA["10006 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="554" y="317" width="130" height="30" uuid="58609327-5433-49fc-a80e-97f046dfa208"/>
				<textFieldExpression><![CDATA["Kundennummer "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s6" x="24" y="367" width="335" height="30" uuid="9734306d-bff4-4e39-a2b4-d132c9a7a47d"/>
				<textFieldExpression><![CDATA["Lieferschein "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="554" y="368" width="60" height="30" uuid="f4717dd5-40cc-4a2e-a374-91efe0abb200"/>
				<textFieldExpression><![CDATA["Seite "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="695" y="368" width="60" height="30" uuid="ad967adc-83a2-4dba-9fab-d18032c1aad5"/>
				<textFieldExpression><![CDATA["1 / 1 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s7" x="24" y="421" width="110" height="29" uuid="3050ea14-ad53-4d2f-93d1-4cc74afa4544"/>
				<textFieldExpression><![CDATA["Versandart "]]></textFieldExpression>
			</textField>
		</band>
	</pageHeader>
	<columnHeader>
		<band height="60" splitType="Stretch">
			<line>
				<reportElement x="25" y="1" width="848" height="1" uuid="74d79a6c-b423-4216-9648-ef215c75eb2b">
					<property name="com.jaspersoft.studio.unit.width" value="px"/>
				</reportElement>
			</line>
			<line>
				<reportElement x="25" y="25" width="848" height="1" uuid="74d79a6c-b423-4216-9648-ef215c75eb2b">
					<property name="com.jaspersoft.studio.unit.width" value="px"/>
				</reportElement>
			</line>
			<textField>
				<reportElement style="t s2" x="73" y="2" width="70" height="30" uuid="8c6cf1fe-5b8f-4cc0-b874-2d7ab02ca255"/>
				<textFieldExpression><![CDATA["Modell "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s2" x="312" y="2" width="60" height="30" uuid="884b6f75-3319-4dfc-a634-bd3e8c6b55bd"/>
				<textFieldExpression><![CDATA["Farbe "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s2" x="30" y="2" width="40" height="30" uuid="ce9d6b00-4d94-4765-a956-ae84ae906ad6"/>
				<textFieldExpression><![CDATA["Pos "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s2" x="818" y="2" width="82" height="30" uuid="26d70d97-e29a-4c54-b970-c9db18f405f8"/>
				<textFieldExpression><![CDATA["Stück "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s2" x="439" y="3" width="80" height="30" uuid="0a40f253-f8a4-4f6b-bbcf-254b81b2475e"/>
				<textFieldExpression><![CDATA["Artikel "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="30" y="28" width="370" height="30" uuid="85a6b313-1e62-4e01-93ec-1246d315e1f9"/>
				<textFieldExpression><![CDATA["PO TT 01/23 WOMEN Zalando PO1416415C "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s0" x="313" y="28" width="110" height="30" uuid="753c8bd2-7c16-4fbb-ae47-edaf6c98c033"/>
				<textFieldExpression><![CDATA["Lodz Auftr "]]></textFieldExpression>
			</textField>
			<line>
				<reportElement x="435" y="25" width="1" height="35" uuid="1acd1808-f1d9-4dba-91fb-4c7f43c408d3"/>
			</line>
			<line>
				<reportElement x="499" y="25" width="1" height="35" uuid="1a15bd76-c24d-4621-9ed5-b67f3678e96e"/>
			</line>
			<line>
				<reportElement x="24" y="1" width="1" height="25" uuid="1297727a-c3c3-4b22-b327-8b527081a823">
					<property name="com.jaspersoft.studio.unit.height" value="px"/>
					<property name="com.jaspersoft.studio.unit.y" value="px"/>
				</reportElement>
			</line>
			<line>
				<reportElement x="873" y="1" width="1" height="25" uuid="b7604b72-1094-4510-8ecd-a85af77e975e"/>
			</line>
		</band>
	</columnHeader>
	<detail>
		<band height="60" splitType="Stretch">
			<textField>
				<reportElement style="t s4" x="607" y="0" width="20" height="30" uuid="e78ea146-bb08-4d9e-a05a-b21487c97e06"/>
				<textFieldExpression><![CDATA["S "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="677" y="0" width="20" height="30" uuid="6e71152e-3b79-4a42-af26-c0beafdb244a"/>
				<textFieldExpression><![CDATA["L "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="766" y="0" width="40" height="30" uuid="25df0ca0-73a8-40a2-8e1e-27219f8428c7"/>
				<textFieldExpression><![CDATA["3XL "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="704" y="0" width="30" height="30" uuid="771aae79-0af6-4648-87a0-2e5b22d481da"/>
				<textFieldExpression><![CDATA["XL "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="731" y="0" width="40" height="30" uuid="10a5e26a-7133-43c3-8d07-dd35eddafb00"/>
				<textFieldExpression><![CDATA["XXL "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="640" y="0" width="20" height="30" uuid="b7db5d44-93d7-4b54-b042-bf5e89c5993f"/>
				<textFieldExpression><![CDATA["M "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="563" y="0" width="30" height="30" uuid="04c24e5d-e78f-4b3d-aacb-c1a427ac722e"/>
				<textFieldExpression><![CDATA["XS "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="521" y="0" width="40" height="30" uuid="d1ca2b90-2b8c-42dd-8197-e0831e58cd42"/>
				<textFieldExpression><![CDATA["XXS "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s3" x="444" y="24" width="50" height="30" uuid="18edbb6b-f36c-4125-a868-06efd49c728e"/>
				<textFieldExpression><![CDATA["2311 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s3" x="312" y="24" width="100" height="30" uuid="6d1d5c44-ac31-4a98-a18f-427953b8a9a8"/>
				<textFieldExpression><![CDATA["TANGERINE "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s3" x="191" y="24" width="120" height="30" uuid="2a4a6700-53ea-4410-9846-4feff88e3c6a"/>
				<textFieldExpression><![CDATA["VALO HOODIE "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="843" y="26" width="40" height="30" uuid="2711f049-dc36-4a13-9252-77ea1175ff66"/>
				<textFieldExpression><![CDATA["119 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="539" y="26" width="20" height="30" uuid="3769b5f3-b1b7-46e4-9a43-e6b90e300886"/>
				<textFieldExpression><![CDATA["1 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="566" y="26" width="30" height="30" uuid="221e4da2-fcf6-4eaa-87d7-f0816db38184"/>
				<textFieldExpression><![CDATA["27 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="601" y="26" width="30" height="30" uuid="c9c4b024-ae04-4735-a053-bde8ea32450c"/>
				<textFieldExpression><![CDATA["47 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="637" y="26" width="30" height="30" uuid="1811b53a-1e9c-4f41-bc41-ef9e55fb0367"/>
				<textFieldExpression><![CDATA["32 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="671" y="26" width="30" height="30" uuid="a558166e-d5d8-402f-8ef4-656ada1e770a"/>
				<textFieldExpression><![CDATA["12 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="73" y="26" width="80" height="30" uuid="e8ac5007-2a06-4842-8695-b706b33e4cfb"/>
				<textFieldExpression><![CDATA["W2023TT "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="48" y="30" width="20" height="30" uuid="5dd42759-ad57-4dca-a008-27ff0895622a"/>
				<textFieldExpression><![CDATA["1 "]]></textFieldExpression>
			</textField>
			<line>
				<reportElement x="25" y="58" width="848" height="1" uuid="74dc085e-84d6-4808-ae37-57961051d2f9">
					<property name="com.jaspersoft.studio.unit.width" value="px"/>
					<property name="com.jaspersoft.studio.unit.x" value="px"/>
					<property name="com.jaspersoft.studio.unit.y" value="px"/>
				</reportElement>
			</line>
			<line>
				<reportElement x="499" y="-2" width="1" height="61" uuid="53778345-5f02-4592-bb25-b6724ccf8073"/>
			</line>
			<line>
				<reportElement x="435" y="-2" width="1" height="61" uuid="a972f60a-944b-4451-80b1-06abc6f32197"/>
			</line>
		</band>
	</detail>
	<columnFooter>
		<band height="45" splitType="Stretch">
			<textField>
				<reportElement style="t s1" x="842" y="0" width="40" height="30" uuid="b3dfd69b-4b7c-4dce-82e0-bca39b26cb58"/>
				<textFieldExpression><![CDATA["272 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s4" x="704" y="0" width="140" height="30" uuid="14fad654-32a5-46dd-b029-6c8b84124ec3"/>
				<textFieldExpression><![CDATA["Gesamt Stück: "]]></textFieldExpression>
			</textField>
		</band>
	</columnFooter>
	<pageFooter>
		<band height="600" splitType="Stretch">
			<textField>
				<reportElement style="t s1" x="24" y="0" width="410" height="30" uuid="226f6248-4a93-429e-8806-ded800f557b6"/>
				<textFieldExpression><![CDATA["Sitz der Gesellschaft / Lieferanschrift: "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="263" y="0" width="160" height="30" uuid="8784c329-4fa6-431d-ba61-725a3f7e59dc"/>
				<textFieldExpression><![CDATA["Bankverbindung: "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="664" y="0" width="180" height="30" uuid="ddc6942c-5d0f-45ac-b412-39989ce1172f"/>
				<textFieldExpression><![CDATA["Geschäftsführung: "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="486" y="0" width="170" height="30" uuid="5af5f189-0d1d-41f3-9a84-9c3747951380"/>
				<textFieldExpression><![CDATA["Amtsgericht: Ulm "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="24" y="15" width="200" height="30" uuid="69ed5bfa-09d3-45cc-828d-1a8539054557"/>
				<textFieldExpression><![CDATA["Worst Behavior GmbH "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="263" y="15" width="240" height="30" uuid="a50f9940-fe12-4e71-af2b-173ca1d1b881"/>
				<textFieldExpression><![CDATA["Kreissparkasse Biberach "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="664" y="15" width="310" height="30" uuid="9e77f31b-5a96-4f47-a3fe-f8af3694fc75"/>
				<textFieldExpression><![CDATA["Fabian Altmann, Janek Feldmann "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="486" y="15" width="280" height="30" uuid="ec39bd3d-1518-4cb0-9772-fb516175e2e8"/>
				<textFieldExpression><![CDATA["Handelsregister: HRB 728233 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="24" y="29" width="200" height="30" uuid="6be1b77a-d233-4c12-b229-d630f456b52c"/>
				<textFieldExpression><![CDATA["Carl-Benz-Straße 17 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="263" y="29" width="290" height="30" uuid="f468cb21-32be-4e5f-adb0-c0960d0c7905"/>
				<textFieldExpression><![CDATA["IBAN: DE19654500700007620287 "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="24" y="43" width="210" height="30" uuid="acd903c3-7d9c-41b3-83f0-946652ae5470"/>
				<textFieldExpression><![CDATA["D-89597 Munderkingen "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="263" y="43" width="170" height="30" uuid="7e52f0f9-d1e1-41ca-8a75-59fe3dcfe095"/>
				<textFieldExpression><![CDATA["BIC: SBCRDE66XXX "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="664" y="43" width="350" height="30" uuid="9d858186-0497-4178-9030-7844f7ee79bd"/>
				<textFieldExpression><![CDATA["E-Mail: support@worst-behavior.com "]]></textFieldExpression>
			</textField>
			<textField>
				<reportElement style="t s1" x="486" y="43" width="220" height="30" uuid="b5db6763-d6d7-4156-9b2c-0bb739f706c6"/>
				<textFieldExpression><![CDATA["Ust-IDNr. DE284749052 "]]></textFieldExpression>
			</textField>
		</band>
	</pageFooter>
	<summary>
		<band height="42" splitType="Stretch"/>
	</summary>
</jasperReport>

    """
# The testing code would look something like this:
print(transform_jasper_report(xml_content))

```

```python
import random

def generate_sql_inserts(num_articles=5, max_headers=13):
    # Centralized definitions
    VALUE_COLUMNS = [f"value{i}" for i in range(max_headers)]
    ADDITIONAL_COLUMNS = {
        "SizeHeader0": "VARCHAR(10)",
        "SizeHeader1": "VARCHAR(10)",
        "SizeHeader2": "VARCHAR(10)",
        "SizeHeader3": "VARCHAR(10)",
        "SizeHeader4": "VARCHAR(10)",
        "SizeHeader5": "VARCHAR(10)",
        "SizeHeader6": "VARCHAR(10)",
        "SizeHeader7": "VARCHAR(10)",
        "ArticleNumValue": "INT",
        "ArticleColorValue": "VARCHAR(20)",
        "ArticleModellNameValue": "VARCHAR(50)",
        "ArticleCountValue": "INT",
        "ArticleSizeValue0": "INT",
        "ArticleSizeValue1": "INT",
        "ArticleSizeValue2": "INT",
        "ArticleSizeValue3": "INT",
        "ArticleSizeValue4": "INT",
        "ArticleSizeValue5": "INT",
        "ArticleSizeValue6": "INT",
        "ArticleSizeValue7": "INT",
        "ArticleModellCode": "VARCHAR(50)",
        "ArticlePosValue": "INT"
    }

    def generate_table_columns():
        basic_columns = ", ".join([f"{col} VARCHAR(10)" for col in VALUE_COLUMNS])
        additional_columns = ", ".join([f"{col} {type}" for col, type in ADDITIONAL_COLUMNS.items()])
        return basic_columns + ", " + additional_columns

    # Table cleanup and creation
    init_statements = [
        "DROP TABLE IF EXISTS ArticleDataRows;",
        "DROP TABLE IF EXISTS Articles;",
        "DROP TABLE IF EXISTS Company;",
        "CREATE TABLE Company (id INT PRIMARY KEY, name VARCHAR(255) NOT NULL);",
        "CREATE TABLE Articles (id INT PRIMARY KEY, article_name VARCHAR(255) NOT NULL, article_rows_data_values_sum INT NOT NULL, article_row_count_max INT NOT NULL, article_row_count INT NOT NULL);",
        f"CREATE TABLE ArticleDataRows (id INT PRIMARY KEY, article_id INT, row_num INT, row_name VARCHAR(10), row_data_values_sum INT NOT NULL, data_count INT NOT NULL, {generate_table_columns()}, FOREIGN KEY (article_id) REFERENCES Articles(id));"
    ]

    company_statements = ["INSERT INTO Company (id, name) VALUES (1, 'CompanyName');"]
    article_statements = []
    row_statements = []
    article_data_row_id = 1  # To keep track of the unique id for ArticleDataRows table

    for article_id in range(1, num_articles + 1):
        article_name = f'Artikel {article_id}'
        article_row_count_max = 3
        article_row_count = random.randint(1, article_row_count_max)
        article_rows_data_values_sum = 0

        for row_num in range(1, article_row_count + 1):
            row_name = f'Row {chr(64 + row_num)}:'
            data_count = random.randint(1, max_headers)
            values = [str(random.randint(0, 29)) for _ in range(data_count)] + ['0'] * (max_headers - data_count)
            row_data_values_sum = sum([int(value) for value in values[:data_count]])
            article_rows_data_values_sum += row_data_values_sum

            additional_values = {
                "SizeHeader0": f"'Size{random.randint(1, 9)}'",
                "SizeHeader1": f"'Size{random.randint(1, 9)}'",
                #... and so on for other SizeHeader fields
                "ArticleNumValue": str(random.randint(1000, 9999)),
                "ArticleColorValue": f"'Color{random.randint(1, 5)}'",
                "ArticleModellNameValue": f"'Model{random.randint(1, 5)}'",
                "ArticleCountValue": str(random.randint(1, 10)),
                "ArticleSizeValue0": str(random.randint(28, 38)),
                #... and so on for other ArticleSizeValue fields
                "ArticleModellCode": f"'Code{random.randint(100, 500)}'",
                "ArticlePosValue": str(random.randint(1, 5))
            }

            all_columns = VALUE_COLUMNS + list(additional_values.keys())
            all_values = values + list(additional_values.values())
            values_str = ", ".join(all_values)

            row_statements.append(
                f"INSERT INTO ArticleDataRows ({', '.join(all_columns)}) VALUES ({article_data_row_id}, {article_id}, {row_num}, '{row_name}', {row_data_values_sum}, {data_count}, {values_str});"
            )
            article_data_row_id += 1

        article_statements.append(
            f"INSERT INTO Articles (id, article_name, article_rows_data_values_sum, article_row_count_max, article_row_count) VALUES ({article_id}, '{article_name}', {article_rows_data_values_sum}, {article_row_count_max}, {article_row_count});"
        )

    return '\n'.join(init_statements + company_statements + article_statements + row_statements)

generated_sql = generate_sql_inserts()
print(generated_sql)
```

```python
# 3. Code-Sektionen

import re
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET
import random

def convert_bottom_to_top(width: int, height: int, convertedtext: str) -> str:
    # Use regex to find all 'bottom' properties and their values
    def replace_bottom_to_top(match):
        bottom_value = float(match.group(1))
        top_value = height - bottom_value
        return f"top:{str(round(top_value))}px"

    # Replace 'bottom' with the calculated 'top' value
    updated_text = re.sub(r'bottom:(\d+\.?\d*)px', replace_bottom_to_top, convertedtext)
    return updated_text

def generate_jasper_textfields_from_html_multi_style(converthtml: str) -> str:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(converthtml, 'html.parser')

    # Extract all styles from all style tags
    styles = {}
    for style_content in soup.select('style'):
        if style_content.string:  # Check if style content is not None
            for style_definition in style_content.string.split('}'):
                if '{' in style_definition:
                    parts = style_definition.split('{')
                    if len(parts) == 2:
                        selector, properties = parts
                        selector = selector.strip()
                        properties = properties.strip()
                        styles[selector] = properties

    # Process each span to extract the text and associated styles
    jasper_textfields = ""
    for span in soup.select('span.t'):
        span_id = span.attrs.get('id', '')
        span_text = span.string
        span_class = " ".join(span.attrs.get('class', []))

        # Extract style associated with this span using its id
        style_definition = styles.get(f"#{span_id}", '')
        x_match = re.search(r'left:(\d+)px', style_definition)
        y_match = re.search(r'top:(\d+)px', style_definition)
        x_pos = x_match.group(1) if x_match else '0' # Default to '0' if not found
        y_pos = y_match.group(1) if y_match else '0' # Default to '0' if not found


        # Generate the JasperReport XML for this text span
        jasper_textfields += f"""
            <textField>
                <reportElement style="{span_class}" x="{x_pos}" y="{y_pos}" width="{10 * len(str(span_text))}" height="30" />
                <textFieldExpression><![CDATA["{span_text}"]]></textFieldExpression>
            </textField>
        """

    return jasper_textfields

def extract_jasper_styles_from_html_styles(html_styles: str) -> str:
    # Regular expression patterns for extracting font family and font size
    font_family_pattern = re.compile(r'font-family:([^;]+)')
    font_size_pattern = re.compile(r'font-size:(\d+)px')

    jasper_styles = ""

    # Split the styles by "}" to get individual style definitions
    for style_definition in html_styles.split('}'):
        # Check if style definition is not empty
        if style_definition.strip():
            # Extract class name (e.g., s0, s1, etc.)
            class_name = style_definition.split('{')[0].strip().replace('.', '')

            # Extract font family and font size
            font_family = font_family_pattern.search(style_definition)
            font_size = font_size_pattern.search(style_definition)

            if font_family and font_size:
                jasper_styles += f'<style name="t {class_name}" fontName="{font_family.group(1).strip()}" fontSize="{font_size.group(1)}"/>\n'

    return jasper_styles

# Re-defining the improved function
def sort_textfields_by_y_etree_v3(xml_content: str) -> str:
    # Placeholder for CDATA
    placeholder = "PLACEHOLDER_FOR_CDATA"

    # Convert CDATA sections to placeholder
    xml_content = xml_content.replace("<![CDATA[", placeholder + "[")
    xml_content = xml_content.replace("]]>", "]" + placeholder)

    # Define namespace
    ns = {'jr': 'http://jasperreports.sourceforge.net/jasperreports'}
    ET.register_namespace('', ns['jr'])

    # Parse the XML content with the namespace
    root = ET.fromstring(xml_content)

    # For each parent in the XML
    for parent in root.findall(".//*"):
        # Extract textField elements and their y coordinates
        textfields = []
        for elem in parent.findall("jr:textField", ns):
             report_element = elem.find(".//jr:reportElement", ns)
             if report_element is not None and 'y' in report_element.attrib:
                 textfields.append((elem, int(report_element.attrib['y'])))


        # If there are no textFields, continue to the next parent
        if not textfields:
            continue

        # Sort the textFields based on y coordinate
        textfields.sort(key=lambda x: x[1])

        # Remove the original textFields from parent
        for tf, _ in textfields:
            parent.remove(tf)

        # Add sorted textField elements back to parent
        for tf, _ in textfields:
            parent.append(tf)

    # Convert the root element back to string and restore CDATA sections
    output = ET.tostring(root, encoding='unicode')
    output = output.replace(placeholder + "[", "<![CDATA[")
    output = output.replace("]" + placeholder, "]]>")

    return output

def adjust_jasperreport_positions_regex(xml_content: str, x_offset=0, y_offset=0) -> str:

    # Define a function to update x and y values using regex
    def repl(match):
        if match.group(1) == "x":
            return 'x="{}"'.format(int(match.group(2)) + x_offset)
        if match.group(1) == "y":
            return 'y="{}"'.format(int(match.group(2)) + y_offset)
        return match.group(0)

    # Apply the regex replacement
    adjusted_xml = re.sub(r'(\b[x|y])="(\d+)"', repl, xml_content)
    return adjusted_xml

def adjust_jasperreport_positions_corrected(xml_content: str, x_scale_factor=1.0, y_scale_factor=1.0, font_scale_factor=1.0, height_scale_factor=1.0, width_scale_factor=1.0) -> str:

    def repl(match):
        attribute = match.group(1)
        value = match.group(2)

        # Scale the value if necessary
        if attribute == "x":
            value = str(round(float(value) * x_scale_factor))
        elif attribute == "y":
            value = str(round(float(value) * y_scale_factor))
        elif attribute == "fontSize":
            value = str(round(float(value) * font_scale_factor))
        elif attribute == "height":
            value = str(round(float(value) * height_scale_factor))
        elif attribute == "width":
            value = str(round(float(value) * width_scale_factor))

        return attribute + '="' + value + '"'

    # Correcting the regex pattern to capture decimal values as well
    adjusted_xml = re.sub(r'(\bx|y|fontSize|height|width)="([\d.]+)"', repl, xml_content)
    return adjusted_xml

def adjust_jasperreport_positions_regex_selective(xml_content: str, x_scale_factor=1.0, y_scale_factor=1.0, font_scale_factor=1.0, height_scale_factor=1.0, width_scale_factor=1.0, isTextFieldScaling=True, scaleAll=False) -> str:

    def repl(match):
        # Check if the current match is inside a <textField> element
        preceding_text = xml_content[:match.start(0)]
        is_inside_textfield = "<textField>" in preceding_text.split("</textField>")[-1]

        attribute = match.group(1)
        value = match.group(2)


        # Determine if the current match should be scaled
        if scaleAll:
            should_scale = True
        else:
            should_scale = (is_inside_textfield and isTextFieldScaling) or (not is_inside_textfield and not isTextFieldScaling)

        # Scale the value if necessary
        if should_scale:
            if attribute == "x":
                value = str(round(int(value) * x_scale_factor))
            elif attribute == "y":
                value = str(round(int(value) * y_scale_factor))
            elif attribute == "fontSize":
                value = str(round(int(value) * font_scale_factor))
            elif attribute == "height":
                value = str(round(int(value) * height_scale_factor))
            elif attribute == "width":
                value = str(round(int(value) * width_scale_factor))

        return attribute + '="' + value + '"'

    adjusted_xml = re.sub(r'(\bx|y|fontSize|height|width)="(\d+)"', repl, xml_content)
    return adjusted_xml

def transform_jasper_report(xml_content: str) -> str:
    added_parameters = []
    added_fields = []

    skip_all = False  # global variable to check if the user has chosen the "Skip All" option

    # Diese Funktion ersetzt das textField anhand der Auswahl des Benutzers
    def replace_text_field(match):
        nonlocal skip_all
        # Wenn skip_all True ist, geben Sie das textField so zurück, wie es ist
        if skip_all:
            return match.group(0)

        # textField-Inhalt extrahieren
        text_field_content = match.group(0)

        # Fragen Sie den Benutzer nach einer Wahl
        print("Für das textField:", text_field_content)
        choice = input("Wählen Sie eine Aktion (1: Parameter [default], 2: Feld, 3: Überspringen, 4: Alles überspringen): ")

        # Wenn keine Auswahl getroffen wird, wird "Parameter" als Standardwert festgelegt
        if choice.strip() == '':
            choice = '1'

        # Wenn der Benutzer "Alles überspringen" wählt
        if choice == '4':
            skip_all = True
            return text_field_content

        # Wenn der Benutzer überspringt
        if choice == '3':
            return text_field_content

        # Extrahieren Sie den String aus textFieldExpression
        expression_match = re.search(r'<textFieldExpression><!\[CDATA\[(.*?)\]\]></textFieldExpression>', text_field_content, re.DOTALL)
        if not expression_match:
            return text_field_content
        extracted_string = expression_match.group(1).strip() # strip whitespace

        # Fragen Sie nach dem Namen
        name = input(f"Geben Sie einen Namen für den {'Parameter' if choice == '1' else 'Feld'} ein: ")

        # Fragen Sie nach dem Datentyp
        data_type = input(f"Wählen Sie den Datentyp (1: java.lang.String [default], 2: java.lang.Integer): ")
        if data_type.strip() == '' or data_type == '1':
            data_type = 'java.lang.String'
        else:
            data_type = 'java.lang.Integer'

        # Erstellen Sie Parameter oder Feld basierend auf der Wahl des Benutzers
        if choice == '1':
            param_str = (
                f'<parameter name="{name}" class="{data_type}">\n'
                f'\t<defaultValueExpression><![CDATA[{extracted_string}]]></defaultValueExpression>\n'
                f'</parameter>'
            )
            added_parameters.append(param_str)
        else:
            field_str = f'<field name="{name}" class="{data_type}">\n\t<defaultValueExpression><![CDATA[{extracted_string}]]></defaultValueExpression>\n</field>'
            added_fields.append(field_str)

        # Erstellen Sie den Ersatzstring
        prefix = 'P' if choice == '1' else 'F'
        replacement = f'<textFieldExpression><![CDATA[${prefix}{{{name}}}]]></textFieldExpression>'

        # Ersetzen Sie den Inhalt mit re.sub()
        new_text_field_content = re.sub(r'<textFieldExpression><!\[CDATA\[(.*?)\]\]></textFieldExpression>', replacement, text_field_content, re.DOTALL)

        return new_text_field_content

    # Replace all textFields in the XML content
    modified_xml = re.sub(r'(<textField>.*?</textField>)', replace_text_field, xml_content, flags=re.DOTALL)

    # Find the insertion point for parameters (before the first <queryString> or at the end of styles)
    params_insertion_point_match = re.search(r'<queryString', modified_xml)
    if params_insertion_point_match:
         params_insertion_point = params_insertion_point_match.start()
         modified_xml = modified_xml[:params_insertion_point] + "\n\n<!-- Added Parameter Definitions -->\n" + "\n".join(added_parameters) + "\n" + modified_xml[params_insertion_point:]
    else:
        # If no queryString, find the end of styles
        styles_end_match = re.search(r'</style>', modified_xml)
        if styles_end_match:
            styles_end_point = styles_end_match.end()
            modified_xml = modified_xml[:styles_end_point] + "\n\n<!-- Added Parameter Definitions -->\n" + "\n".join(added_parameters) + "\n" + modified_xml[styles_end_point:]
        else:
             # If no styles or query string, add at the end of the report tag
             report_end_match = re.search(r'</jasperReport>', modified_xml)
             if report_end_match:
                report_end_point = report_end_match.start()
                modified_xml = modified_xml[:report_end_point] + "\n\n<!-- Added Parameter Definitions -->\n" + "\n".join(added_parameters) + "\n" + modified_xml[report_end_point:]
             else:
                modified_xml += "\n\n<!-- Added Parameter Definitions -->\n"
                modified_xml += "\n".join(added_parameters)


    # Find the insertion point for fields (after the </queryString>)
    fields_insertion_point_match = re.search(r'</queryString>', modified_xml)
    if fields_insertion_point_match:
        fields_insertion_point = fields_insertion_point_match.end()
        modified_xml = modified_xml[:fields_insertion_point] + "\n\n<!-- Added Field Definitions -->\n" + "\n".join(added_fields) + "\n" + modified_xml[fields_insertion_point:]
    else:
        # If no queryString, add after the parameters
        modified_xml += "\n\n<!-- Added Field Definitions -->\n"
        modified_xml += "\n".join(added_fields)


    return modified_xml


def generate_sql_inserts_articles(num_articles=5, max_headers=13):
    # Centralized definitions
    VALUE_COLUMNS = [f"value{i}" for i in range(max_headers)]
    ADDITIONAL_ARTICLE_COLUMNS = {
        "ArticleNumValue": "INT",
        "ArticleColorValue": "VARCHAR(20)",
        "ArticleModellNameValue": "VARCHAR(50)",
        "ArticleCountValue": "INT",
        "ArticleModellCode": "VARCHAR(50)",
    }

    def generate_article_table_columns():
        return ", ".join([f"{col} {type}" for col, type in ADDITIONAL_ARTICLE_COLUMNS.items()])


    # Table cleanup and creation
    init_statements = [
        "DROP TABLE IF EXISTS ArticleDataRows;",
        "DROP TABLE IF EXISTS Articles;",
        "DROP TABLE IF EXISTS Company;",
        "CREATE TABLE Company (id INT PRIMARY KEY, name VARCHAR(255) NOT NULL);",
        f"CREATE TABLE Articles (id INT PRIMARY KEY, article_name VARCHAR(255) NOT NULL, article_rows_data_values_sum INT NOT NULL, article_row_count_max INT NOT NULL, article_row_count INT NOT NULL, {generate_article_table_columns()});",
        f"CREATE TABLE ArticleDataRows (id INT PRIMARY KEY, article_id INT, row_num INT, row_name VARCHAR(10), row_data_values_sum INT NOT NULL, data_count INT NOT NULL, {', '.join([f'{col} VARCHAR(10)' for col in VALUE_COLUMNS])}, FOREIGN KEY (article_id) REFERENCES Articles(id));"
    ]


    company_statements = ["INSERT INTO Company (id, name) VALUES (1, 'CompanyName');"]
    article_statements = []
    row_statements = []
    article_data_row_id = 1  # To keep track of the unique id for ArticleDataRows table

    for article_id in range(1, num_articles + 1):
        article_name = f'Artikel {article_id}'
        article_row_count_max = 3
        article_row_count = random.randint(1, article_row_count_max)
        article_rows_data_values_sum = 0

        # Generate additional article values
        additional_article_values = {
             "ArticleNumValue": str(random.randint(1000, 9999)),
             "ArticleColorValue": f"'Color{random.randint(1, 5)}'",
             "ArticleModellNameValue": f"'Model{random.randint(1, 5)}'",
             "ArticleCountValue": str(random.randint(1, 10)),
             "ArticleModellCode": f"'Code{random.randint(100, 500)}'",
         }
        article_columns_str = ", ".join(list(ADDITIONAL_ARTICLE_COLUMNS.keys()))
        article_values_str = ", ".join(list(additional_article_values.values()))

        for row_num in range(1, article_row_count + 1):
            row_name = f'Row {chr(64 + row_num)}:'
            data_count = random.randint(1, max_headers)
            values = [str(random.randint(0, 29)) for _ in range(data_count)] + ['NULL'] * (max_headers - data_count)
            row_data_values_sum = sum([int(value) for value in values if value != 'NULL'])
            article_rows_data_values_sum += row_data_values_sum


            values_str = ", ".join(values)

            row_statements.append(
                f"INSERT INTO ArticleDataRows (id, article_id, row_num, row_name, row_data_values_sum, data_count, {', '.join(VALUE_COLUMNS)}) VALUES ({article_data_row_id}, {article_id}, {row_num}, '{row_name}', {row_data_values_sum}, {data_count}, {values_str});"
            )
            article_data_row_id += 1

        article_statements.append(
            f"INSERT INTO Articles (id, article_name, article_rows_data_values_sum, article_row_count_max, article_row_count, {article_columns_str}) VALUES ({article_id}, '{article_name}', {article_rows_data_values_sum}, {article_row_count_max}, {article_row_count}, {article_values_str});"
        )


    return '\n'.join(init_statements + company_statements + article_statements + row_statements)

def generate_sql_inserts_article_data_rows(num_rows=50):
    inserts = []

    # Centralized definitions
    VALUE_COLUMNS = [f"value{i}" for i in range(13)]
    ADDITIONAL_COLUMNS = {
        "SizeHeader0": "VARCHAR(10)",
        "SizeHeader1": "VARCHAR(10)",
        "SizeHeader2": "VARCHAR(10)",
        "SizeHeader3": "VARCHAR(10)",
        "SizeHeader4": "VARCHAR(10)",
        "SizeHeader5": "VARCHAR(10)",
        "SizeHeader6": "VARCHAR(10)",
        "SizeHeader7": "VARCHAR(10)",
        "ArticleNumValue": "INT",
        "ArticleColorValue": "VARCHAR(20)",
        "ArticleModellNameValue": "VARCHAR(50)",
        "ArticleCountValue": "INT",
        "ArticleSizeValue0": "INT",
        "ArticleSizeValue1": "INT",
        "ArticleSizeValue2": "INT",
        "ArticleSizeValue3": "INT",
        "ArticleSizeValue4": "INT",
        "ArticleSizeValue5": "INT",
        "ArticleSizeValue6": "INT",
        "ArticleSizeValue7": "INT",
        "ArticleModellCode": "VARCHAR(50)",
        "ArticlePosValue": "INT"
    }


    for i in range(num_rows):
        # Generate random values for the columns
        id_value = i + 1
        article_id = random.randint(1, 1)  # Assuming we have articles from 1 to 1
        row_num = random.randint(1, 5)  # Assuming maximum 5 rows per article
        row_name = f'Row {chr(64 + row_num)}:'
        data_count = random.randint(1, 13)
        values = [str(random.randint(0, 1234)) for _ in range(data_count)] + ['NULL'] * (13 - data_count)
        row_data_values_sum = sum([int(value) for value in values if value != 'NULL'])

        size_headers = [f"'XL{random.randint(1, 9)}'" for _ in range(8)]
        article_num_value = random.randint(1000, 9999)
        article_color_value = f"'Color{random.randint(1, 5)}'"
        article_model_name_value = f"'Model{random.randint(1, 5)}'"
        article_count_value = random.randint(1, 10)
        article_size_values = [str(random.randint(60, 1234)) for _ in range(8)]
        article_model_code = f"'Code{random.randint(100, 500)}'"
        article_pos_value = i + 1

        # Construct the INSERT statement
        insert_statement = f"INSERT INTO ArticleDataRows (id, article_id, row_num, row_name, row_data_values_sum, data_count, {', '.join(VALUE_COLUMNS)}, {', '.join(ADDITIONAL_COLUMNS.keys())}) VALUES ({id_value}, {article_id}, {row_num}, '{row_name}', {row_data_values_sum}, {data_count}, {', '.join(values)}, {', '.join(size_headers)}, {article_num_value}, {article_color_value}, {article_model_name_value}, {article_count_value}, {', '.join(article_size_values)}, {article_model_code}, {article_pos_value});"

        inserts.append(insert_statement)

    return '\n'.join(inserts)
```
